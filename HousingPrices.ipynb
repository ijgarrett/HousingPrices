1. longitude: A measure of how far west a house is; a higher value is farther west

2. latitude: A measure of how far north a house is; a higher value is farther north

3. housingMedianAge: Median age of a house within a block; a lower number is a newer building

4. totalRooms: Total number of rooms within a block

5. totalBedrooms: Total number of bedrooms within a block

6. population: Total number of people residing within a block

7. households: Total number of households, a group of people residing within a home unit, for a block

8. medianIncome: Median income for households within a block of houses (measured in tens of thousands of US Dollars)

9. medianHouseValue: Median house value for households within a block (measured in US Dollars)

10. oceanProximity: Location of the house w.r.t ocean/sea

```python
from sklearn.model_selection import train_test_split

X = data.drop(["median_house_value"], axis = 1)
y = data["median_house_value"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
train_data = X_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize = (15, 8))
```

```python
plt.figure(figsize = (15, 8))
sns.heatmap(train_data.corr(numeric_only = True), annot = True, cmap = "YlGnBu")
```

```python
# these 4 features are right skewed, so the log pulls in the large values
# now the distributions resemble more symmetric bell curve
train_data["total_rooms"] = np.log(train_data["total_rooms"] + 1) # avoid log(0)
train_data["total_bedrooms"] = np.log(train_data["total_bedrooms"] + 1)
train_data["population"] = np.log(train_data["population"] + 1)
train_data["households"] = np.log(train_data["households"] + 1)
```

```python
train_data.hist(figsize = (15, 8))
```

```python
# ocean proximity is important, but not numbers, so we will OneHot value encode them
train_data.ocean_proximity.value_counts()
```

```python
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity, dtype = int)).drop(["ocean_proximity"], axis = 1)
```

```python
train_data
```

```python
# new heatmap with the ocean proximity
plt.figure(figsize = (15, 8))
sns.heatmap(train_data.corr(numeric_only = True), annot = True, cmap = "YlGnBu")
```

```python
plt.figure(figsize = (15, 8))
# houses near the coast seem more expensive!
sns.scatterplot(x = "latitude", y = "longitude", data = train_data, hue = "median_house_value", palette = "coolwarm")
```

```python
# number of bedrooms per total rooms
train_data["bedroom_ratio"] = train_data["total_bedrooms"] / train_data["total_rooms"]
# number of rooms per household
train_data["household_rooms"] = train_data["total_rooms"] / train_data["households"]
```

```python
# new heatmap with bedroom_ratio and household_rooms
plt.figure(figsize = (15, 8))
sns.heatmap(train_data.corr(numeric_only = True), annot = True, cmap = "YlGnBu")
```

```python
# linear regression
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train, y_train = train_data.drop(["median_house_value"], axis = 1), train_data["median_house_value"]
X_train_s = scaler.fit_transform(X_train)

reg = LinearRegression()
reg.fit(X_train_s, y_train)
```

```python
# we have to do the scaling on rooms, bedrooms, population, and households on the test data
# we also have to add the ocean features and the feature engineering ones on the test data
test_data = X_test.join(y_test)

test_data["total_rooms"] = np.log(test_data["total_rooms"] + 1)  # avoid log(0)
test_data["total_bedrooms"] = np.log(test_data["total_bedrooms"] + 1)
test_data["population"] = np.log(test_data["population"] + 1)
test_data["households"] = np.log(test_data["households"] + 1)

# one-hot encode ocean_proximity
test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity, dtype=int)).drop(["ocean_proximity"], axis=1)

# number of bedrooms per total rooms
test_data["bedroom_ratio"] = test_data["total_bedrooms"] / test_data["total_rooms"]
# number of rooms per household
test_data["household_rooms"] = test_data["total_rooms"] / test_data["households"]

# ensure test_data has the same columns as training data
test_data = test_data.reindex(columns=train_data.columns, fill_value=0)
```

```python
X_test, y_test = test_data.drop(["median_house_value"], axis = 1), test_data["median_house_value"]
```

```python
X_test_s = scaler.transform(X_test)
```

```python
# Measures the proportion of variance in the target (y_test) that’s explained by the features (X_test) using our model.
reg.score(X_test_s, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(X_train_s, y_train)
```

```python
forest.score(X_test_s, y_test)
```

```python
from sklearn.model_selection import GridSearchCV

# new forest
forest = RandomForestRegressor()

param_grid = {"n_estimators": [100, 200, 300], 
              "min_samples_split": [2, 4], 
              "max_depth": [None, 4, 8]
             }

# GridSearchCV runs exhaustive search over specified hyperparameter values for an estimator (here, 'forest')
# param_grid: dictionary specifying which hyperparameter values to try (all combinations)
# cv=5: use 5-fold cross-validation to evaluate each hyperparameter combo
# scoring="neg_mean_squared_error": use negative mean squared error as evaluation metric (higher is better)
# return_train_score=True: also keep track of training scores to check for overfitting
grid_search = GridSearchCV(forest, param_grid, cv = 5, scoring = "neg_mean_squared_error", return_train_score = True)
grid_search.fit(X_train_s, y_train)
```

```python
best_forest = grid_search.best_estimator_
```

```python
print(best_forest)
```

```python
best_forest.score(X_test_s, y_test)
```
